# Seaborn

--> Built on the top of Matplotlib.

--> It is specially meant for statistical plotting.

--> used if one has to summarize data in visualizations. 

--> More integrated for working with Pandas data frames.


```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```
