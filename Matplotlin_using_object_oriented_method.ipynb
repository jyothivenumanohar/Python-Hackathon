```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
x = range(0, 10)
y = range(0, 10)

# creating a figure object

fig1=plt.figure()

axes=fig1.add_axes([0.1,0.1,0.9,0.9])

axes.plot(x,y)

axes.set_xlabel("x",fontsize=20)

axes.set_ylabel("y",fontsize=20)
#axs.set_title('Negative correlation')
axes.set_title("Graph")
#plt.show()
```

```python
fig2=plt.figure()

fig2_main=fig2.add_axes([0.1,0.1,0.9,0.9])

fig2_inset=fig2.add_axes([0.2,0.6,0.3,0.3])
fig2_main.plot(x,y)
fig2_inset.plot(y,x)
```

# Subplotting on different axes

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(0, 10, 0.1)
y = np.sin(np.pi * x) + x
 
fig, ax = plt.subplots(1)

ax.plot(x, y)

ax.set_xlabel("x",fontsize=40)

ax.set_ylabel("y",fontsize=40)
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(0, 10, 0.1)
y = np.sin(np.pi * x) + x
 
fig, ax = plt.subplots(3)

ax[0].plot(x, y)
ax[1].plot(x,y)
ax[2].plot(y,x)

ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[1].set_xlabel("x1")
ax[1].set_ylabel("y1")

ax[2].set_xlabel("x2")
ax[2].set_ylabel("y2")

plt.tight_layout()
```

```python
fig1,axes1=plt.subplots(nrows=2,ncols=1)

fig2,axes2=plt.subplots(nrows=1,ncols=2)
axes1[0].plot(x,y)
axes1[1].plot(x,y)
```
